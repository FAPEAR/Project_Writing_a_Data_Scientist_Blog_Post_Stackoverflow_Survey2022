# Clean Data

```python
df.info()
```

```python
#Determine by columns the number of null data.
print(df.isnull().sum())
```

```python
#Clear two columns that did not have more key information. 
df = df.loc[df.WebframeHaveWorkedWith.isnull()]
```

```python
df= df[df.NEWCollabToolsHaveWorkedWith.isnull()]
```

```python
#Review the impact of cleaning two columns in general on the rest of the defined columns in relation to null data.
df.info() 
```

```python
#Remove the columns as the final part of the data cleansing process.
df= df.drop(['WebframeHaveWorkedWith', 
            'NEWCollabToolsHaveWorkedWith'],axis=1)
```

# Analyze, Model and Visualize Data (Descriptive Techniques)

First question

```python
#Relationship between the number of participants in the clean database and the classification of educational levels.
plt.xticks(rotation=90)
sns.countplot(data=df, y ='EdLevel', color="red")
plt.suptitle("At what educational level are the developers who participated in the survey?".title(), y = 1, fontsize = 14, weight = "bold");
```

```python
#Annual average in USD by level of education.
df_EdLevel = df.loc[:, ['EdLevel','ConvertedCompYearly']].groupby(by='EdLevel').mean().sort_values('ConvertedCompYearly', ascending=False)
df_EdLevel.head(20)
```

Conclusion: The main level of education of the respondents is Bachelor's degree followed by Master's degree and Secondary school. After these levels are the respondents who did not finish the Bachelor's degree training.

Second question

```python
# What active survey participants do.
plt.xticks(rotation=90)
sns.countplot(data=df.head(600), y ='Employment', color="green")
plt.suptitle("What do the people who participated in the survey do?".title(), y = 1, fontsize = 14, weight = "bold");
```

```python
#Annual average in USD by main occupation activity.
df_Employment = df.loc[:, ['Employment','ConvertedCompYearly']].groupby(by='Employment').mean().sort_values('ConvertedCompYearly', ascending=False)
df_Employment.head(5)
```

Conclusion: The two main activities of the respondents are: Full-time employee and full-time student. The next relevant level is not employed and looking for work.

Third question

```python
#Main branch classification of survey participants.
plt.xticks(rotation=90)
sns.countplot(data=df, y ='MainBranch', color="yellow")
plt.suptitle("What is the main branch of the survey participants?".title(), y = 1, fontsize = 14, weight = "bold");
```

Conclusion: It is not possible to give a correct answer to this question because the respondents were not represented in the response options and for this reason more than 48% concentrated on the None of these options.

Fourth question

```python
#Operating system professional used by the people who completed the survey.
plt.xticks(rotation=90)
sns.countplot(data=df.head(600), y ='OpSysProfessional use', color="blue")
plt.suptitle("What is the operating system professional used by the survey participants?".title(), y = 1, fontsize = 14, weight = "bold");
```

```python
#Operating system personal used by the people who completed the survey.
plt.xticks(rotation=90)
sns.countplot(data=df.head(600), y ='OpSysPersonal use', color="blue")
plt.suptitle("What is the operating system personal used by the survey participants?".title(), y = 1, fontsize = 14, weight = "bold");
```

Conclusion: The main operating system for both personal and professional use is Windows followed by Linux-based.

Fifth question

```python
#Main age ranges of survey participants.
plt.xticks(rotation=90)
sns.countplot(data=df, y ='Age', color="orange")
plt.suptitle("How old are those who participated in the survey?".title(), y = 1, fontsize = 14, weight = "bold");
```

```python
# Main age ranges and their relationship with the average annual salary in USD.
df_Age = df.loc[:, ['Age','ConvertedCompYearly']].groupby(by='Age').mean().sort_values('ConvertedCompYearly', ascending=False)
df_Age.head(6)
```

Conclusion: The first two age ranges concentrate more than 60% of the respondents, thus: - 23 to 34 years of age and 18 to 24 years of age.

Sixth question

```python
#Distribution of survey participants by gender classification where they were identified.
plt.xticks(rotation=90)
sns.countplot(data=df, y ='Gender', color="black")
plt.suptitle("In what gender do those who participated in the survey identify themselves?".title(), y = 1, fontsize = 14, weight = "bold");
```

```python
# Gender classification with its average annual salary allocation in USD.
df_Gender = df.loc[:, ['Gender','ConvertedCompYearly']].groupby(by='Gender').mean().sort_values('ConvertedCompYearly', ascending=False)
df_Gender.head(6)
```

Conclusion: 70% is concentrated in the male gender and 10% in the female gender.

Seventh question

```python
#Taking the original database, the survey participants are classified by country.
df_Stackoverflow_Survey2022['Country'].value_counts()
```

```python
# Taking the original database, Graph of countries with more than 1000 participants in the survey.
Countries = df_Stackoverflow_Survey2022.groupby('Country').filter(lambda x: len(x)>1000)

Countries['Country'].value_counts().plot(kind = 'barh')
plt.title('Countries where most participated in the survey')
plt.xlabel('Count')
plt.ylabel('Countries');
```

```python
#Taking the original database,the 50 countries with the highest annual average in USD, regardless of educational level or experience.
df_Country = df_Stackoverflow_Survey2022.loc[:, ['Country','ConvertedCompYearly']].groupby(by='Country').mean().sort_values('ConvertedCompYearly', ascending=False)
df_Country.head(50)
```

Conclusion: 55% of those surveyed are from the United States, followed by India at 18% and Germany at 14% and compared to the annual average income in USD in developed countries, in order to analyze the income, it is should take into account the cost of living in each country to determine the actual net income.
